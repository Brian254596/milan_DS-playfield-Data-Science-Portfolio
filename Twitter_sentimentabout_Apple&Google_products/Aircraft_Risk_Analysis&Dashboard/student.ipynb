## 3 Data Preperation

```python
# Drop missing values in Make, Model, Air.dammage, purpose.of.flight
df.dropna(subset=['Make', 'Model', 'Aircraft.damage', 'Purpose.of.flight', 'Engine.Type'], inplace=True)
```

```python
# Fill miising values with zero in the injury columns 
injury_columns = [
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries',
    'Total.Uninjured'
]

df[injury_columns] = df[injury_columns].fillna(0)
```

```python
# Ensure the column is string type and strip spaces
df['Purpose.of.flight'] = df['Purpose.of.flight'].astype(str).str.strip()

# Filter out rows where Purpose.of.flight is '0' or 'Unknown'
df = df[~df['Purpose.of.flight'].isin(['0', 'Unknown'])]
```

```python
# Ensure the column is string type and strip spaces
df['Engine.Type'] = df['Engine.Type'].astype(str).str.strip()

# Filter out rows where Engine.Type is '0' or 'Unknown'
df = df[~df['Engine.Type'].isin(['0', 'Unknown'])]
```

```python
# Check the data type of the injury columns
df[injury_columns].dtypes
```

```python
# Change the injury columns into an integer
df[injury_columns] = df[injury_columns].astype(int)
```

```python
# check the data type for this columns  Make, Model, Air.dammage, purpose.of.flight
df[['Make', 'Model', 'Aircraft.damage', 'Purpose.of.flight', 'Engine.Type']].dtypes
```

```python
df[['Make', 'Model', 'Aircraft.damage', 'Purpose.of.flight', 'Engine.Type']] = df[['Make', 'Model', 'Aircraft.damage', 'Purpose.of.flight', 'Engine.Type']].astype(str)
```

```python
# Check for missing values (count and percentage) in selected columns
missing_count = df[selected_columns].isnull().sum()
missing_percent = df[selected_columns].isnull().mean() * 100

# Combine into a single DataFrame
missing_summary = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Percentage': missing_percent
})

# Display the result
missing_summary
```

```python
# Calculate Fatality rate and create a new column ( Fatality rate = [Total Fatal Injuries] / [Total Aboard] )

df['Total.Aboard'] = (
    df['Total.Fatal.Injuries'] +
    df['Total.Serious.Injuries'] +
    df['Total.Minor.Injuries'] +
    df['Total.Uninjured']
)

# Calculate Fatality Rate
df['Fatality.Rate'] = df['Total.Fatal.Injuries'] / df['Total.Aboard']
```

```python
# Combine Make and Model into a new column Make_Model
df['Make_Model'] = df['Make'] + ' ' + df['Model']
```

```python
# Create a new column 'Fatalities' that is equal to 'Total.Fatal.Injuries'
df['Fatalities'] = df['Total.Fatal.Injuries']
```

```python
# Check if the columns were added
df.columns
```

## Data Visualization
  * Create bar chart of Make_Model and fatalities

```python
# Group by 'Make_Model' and calculate the mean of 'Fatalities'
dangerous_aircraft = df.groupby('Make_Model')['Fatalities'].mean()

# Sort by Fatalities in descending order to get the most dangerous (highest fatalities)
top_dangerous_aircraft = dangerous_aircraft.sort_values(ascending=False).head(10)
# Print the top 10 most dangerous aircraft
print(top_dangerous_aircraft)

```

```python
# Create a bar chart to visualize the top 10 most dangerous aircraft with average fatalities on y-axis
plt.figure(figsize=(10, 6))
top_dangerous_aircraft.plot(kind='bar', color='red')
plt.ylabel('Average Fatalities')
plt.xlabel('Aircraft Model')
plt.title('Top 10 Most Dangerous Aircraft Based on Average Fatalities')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by 'Make_Model' and calculate the mean of 'Fatalities'
safe_aircraft = df.groupby('Make_Model')['Fatalities'].mean()

# Sort by Fatalities in ascending order to get the safest (lowest fatalities)
top_safe_aircraft = safe_aircraft.sort_values(ascending=True).head(10)
print(top_safe_aircraft)

```

```python
# Group by purpose and sum fatalities
flight_fatalities = df.groupby('Purpose.of.flight')['Total.Fatal.Injuries'].sum().sort_values(ascending=False)

# Calculate percentages
total_fatalities = flight_fatalities.sum()
percentages = (flight_fatalities / total_fatalities) * 100

# Create pie chart without labels directly on wedges
plt.figure(figsize=(10, 8))
wedges, texts = plt.pie(flight_fatalities, startangle=140)

# Create legend with labels and percentages
legend_labels = [f"{label}: {pct:.1f}%" for label, pct in zip(flight_fatalities.index, percentages)]
plt.legend(wedges, legend_labels, title="Purpose of Flight", loc="center left", bbox_to_anchor=(1, 0.5))

plt.title('Total Fatalities by Purpose of Flight')
plt.axis('equal')  # Ensure pie is circular
plt.tight_layout()
plt.show()
```

```python
df_grouped = df.groupby('Engine.Type')['Fatality.Rate'].mean().reset_index()

# Sorting by Fatality.Rate in descending order (if you want to see highest fatality rate first)
df_grouped = df_grouped.sort_values(by='Fatality.Rate', ascending=False)

# Display the result
df_grouped
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['Engine.Type'], df_grouped['Fatality.Rate'], color='skyblue')

# Adding titles and labels
plt.title('Average Fatality Rate by Engine Type', fontsize=14)
plt.xlabel('Engine Type', fontsize=12)
plt.ylabel('Average Fatality Rate', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
